```python
# With respect to each ratings assigning it into group of categories

ratings_ages = { 'TV-PG': 'Older Kids', 'TV-MA': 'Adults', 'TV-Y7-FV': 'Older Kids', 'TV-Y7': 'Older Kids', 'TV-14': 'Teens', 'R': 'Adults', 'TV-Y': 'Kids', 'NR': 'Adults', 'PG-13': 'Teens', 'TV-G': 'Kids', 'PG': 'Older Kids', 'G': 'Kids', 'UR': 'Adults', 'NC-17': 'Adults' }
df_hypothesis['target_ages'] = df_hypothesis['rating'].replace(ratings_ages)

#let's see unique target ages

df_hypothesis['target_ages'].unique()
```

```python
# Another category target_ages
df_hypothesis['target_ages'] = pd.Categorical(df_hypothesis['target_ages'], categories=['Kids', 'Older Kids', 'Teens', 'Adults'])

# From duration feature extractin string part and after extracting Changing the object type to numeric

df_hypothesis['duration']= df_hypothesis['duration'].str.extract('(\d+)')
df_hypothesis['duration'] = pd.to_numeric(df_hypothesis['duration'])

df_hypothesis.head()
```

```python
# Group by duration and target_ages

group_by_= df_hypothesis[['duration','target_ages']].groupby(by='target_ages')
# Mean of group_by variable
group=group_by_.mean().reset_index()
group
```

```python
# In A and B variable grouping values
A= group_by_.get_group('Kids')
B= group_by_.get_group('Older Kids')
#mean and std. calutation for kids and older kids variables
M1 = A.mean(numeric_only=True)
S1 = A.std(numeric_only=True)
M2= B.mean(numeric_only=True)
S2 = B.std(numeric_only=True)
print('Mean for movies rated for Kids {} \n Mean for movies rated for older kids {}'.format(M1,M2))
print('Std for movies rated for Older Kids {} \n Std for movies rated for kids {}'.format(S2,S1))
```

```python
from scipy.stats import ttest_ind
ttest,pval = ttest_ind(M1, M2)
print(pval)
if pval<0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```

```python
from scipy import stats

# length of groups and DOF(Degree of freedom)
n1 = len(A)
n2= len(B)
print(n1,n2)
dof = n1+n2-2
print('dof',dof)
sp_2 = ((n2-1)*S1**2 + (n1-1)*S2**2) / dof
print('SP_2 =',sp_2)
sp = np.sqrt(sp_2)
print('SP',sp)

# tvalue
t_val = (M1-M2)/(sp * np.sqrt(1/n1 + 1/n2))
print('tvalue',t_val[0])
```

```python
# t-distribution
stats.t.ppf(0.025,dof)
```

## Because the t-value is not in the range, the null hypothesis is rejected.
## As a result, movies rated for kids and older kids are not lasts atleast two hours long.

# **Modeling**

## **1. K-Means Clustering**



```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Using TF-IDF vectorizer to convert text data to numerical data
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['description'])
```

```python
wcss_list = []  # Initializing the list for the values of WCSS

# Using for loop for iterations from 1 to 30.
for i in range(1, 30):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)

# Plotting the elbow method graph
plt.plot(range(1, 30), wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

# Insight
The Elbow Method graph is used to determine the optimal number of clusters (k) in a K-Means clustering analysis. The point where the decrease in within-cluster sum of squares (WCSS) starts to slow down is often considered as the optimal number of clusters. First it is decreasing near 5 so, the optimal no. of cluster is 5 .

```python
from sklearn.metrics import silhouette_score

#sillhoute score of clusters
sill = []
for i in range(2,30):
  model = KMeans(n_clusters=i,init ='k-means++',random_state=51)
  model.fit(X)
  y1 = model.predict(X)
  score = silhouette_score(X,y1)
  sill.append(score)
  print('cluster: %d \t Sillhoute: %0.4f'%(i,score))
# Plotting the silhouette scores
plt.plot(range(2, 30), sill, marker='o')
plt.title('Silhouette Scores for Different Cluster Numbers')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

# Insight
A silhouette score of 0.0075 is relatively low and suggests that the clusters in data may overlap or be poorly defined.

```python
#training the K-means model on a dataset
kmeans = KMeans(n_clusters= 26, init='k-means++', random_state= 42)
y_predict= kmeans.fit_predict(X)
```

### **Evaluation**

```python
# Predict the clusters and evaluate the silhouette score
score = silhouette_score(X, y_predict)
print("Silhouette score is {}".format(score))
```

```python
#davies_bouldin_score of our clusters
from sklearn.metrics import davies_bouldin_score
davies_bouldin_score(X.toarray(), y_predict)
```

# Insight
A lower Davies-Bouldin Index is desirable, indicating better-defined and more separated clusters. In this case, the high score suggests that the clusters in data may not be well-separated or have clear boundaries.
